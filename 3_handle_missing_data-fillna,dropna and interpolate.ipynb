### Topics Covered In this
1. Fillna to fill missing values using different ways
2. Interpolate to make a guess on missing values using interpolation
3. dropna to drop rows with missing values

```python
import pandas as pd
df=pd.read_csv('weather_data.csv',parse_dates=['day'])#we converted day type (str) to date
type(df['day'][0])
df
df.set_index('day',inplace=True)
df
```

```python
new_df=df.fillna(0)
new_df
```

```python
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
    
})
new_df
```

```python
new_df=df.fillna(method="ffill")
new_df
#by this method we carry forwarded the upper values to blank method 
```

```python
new_df=df.fillna(method='bfill')
new_df
#it will fill down cell value to upper cell value where value is 0
#here it is filling value vertically
```

```python
#in above cell  we filling value vertically
#now we are filling value horizontally by axis=colums
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

```python
new_df=df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df=df.interpolate()
new_df
#it is using linear interpolation
#you can also use many method such as quadric,cubic,polynomial,time
```

```python
#method=time
new_df=df.interpolate(method='time')
new_df
```

```python
new_df=df.dropna()
new_df#it will drop all the missing values
```

```python
#i want to remove all which have all missing value
new_df=df.dropna(how="all")
new_df
```

```python
new_df=df.dropna(thresh=2)#thresh =2 means i want two valid value to keep that row
new_df
```

```python
#if i have some missing date and i want to fill that
dt=pd.date_range("01/01/2017","01/11/2017")
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
