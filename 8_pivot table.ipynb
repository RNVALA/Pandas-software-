1. Pivot allows you to transform or reshape data

```python
import pandas as pd
df=pd.read_csv('pivot weather 1.csv')
df
```

```python
df.pivot(index='date',columns='city')
# first argument what you want in x axis  or row of data
# second argument is columns
```

```python
df.pivot(index='date',columns='city',values='humidity')
# first argument what you want in x axis  or row of data
# second argument is columns
#let take humidity is present in my output data frame
```

```python
df.pivot(index='humidity',columns='city')
```

#### What is Pivot table
1.) Pivot table is used to summarize and aggregate data inside dataframe

```python
df=pd.read_csv('pivot weather 2.csv')
df
#this is a table in which we can see that both data is there morning weather and evening weather
#and we require  average of humidity and temperature 
```

```python
df.pivot_table(index='city',columns='date')# wecan add aggfunc='sum' etc
#default aggfun is mean
```

```python
df.pivot_table(index='city',columns='date',margins=True)
```

```python
df=pd.read_csv('pivot weather 3.csv')
df
```
