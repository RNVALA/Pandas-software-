```python
import pandas as pd
df1=pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]

})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','orlando','chicago'],
    'humidity':[65,38,75]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',indicator=True)#by default it is outer join
df3
#it will take only inner join or intersection
#here indicator will tell you that from where the values are coming
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')#here we supply outer join
df3
```

```python
#suffix
df1=pd.DataFrame({
    'city':['chicango','new york','orlando','san digo'],
    'temparature':[21,14,35,38],
    'humidity':[65,68,71,75]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['chicango','new york','san diego'],
    'temperature':[21,14,35],
    'humidity':[65,68,71]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',suffixes=('_left','_right'),how='outer',indicator=True)
df3
```
