```python
import pandas as pd
import numpy as np
df=pd.read_csv('weather_data_6.csv')
df
```

```python
new_df=df.replace(-99999,np.NaN)
new_df
#replalcing a value  -99999 withh nan
```

```python
new_df=df.replace([-99999,-88888],np.NaN)
new_df
#replacing two
```

```python
#replace data with specific columns
new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'
},np.NaN)
new_df
```

```python
#simply mapping with values
new_df=df.replace({
    -99999:np.NaN,
    0:'Sunny'
})
new_df
```
